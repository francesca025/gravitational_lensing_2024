# Critical & Caustic Curves of a Source behind a SIS Lens

Demonstrating the critical curves (lens plane) & corresponding caustic curves (source plane) arising from viewing a source from behind a Single Isothermal Sphere (SIS) gravitational lens. 

```python
#import packages, set formatting
from numpy import *
from matplotlib.pyplot import *
rcParams.update({'font.family':'serif','font.size':14,'text.usetex':False})
```

```python
#define data

# Image positions
xi=[-0.35,-0.35,0.86,-0.7]
yi=[0.53,-0.53,0.0,0.0]

# Source position
us=0.07
vs=0.0

# Critical radius
b=0.7

# Shear
gamma=0.1

# Parametrizing variable
dphi=pi/1000.0
phi=arange(dphi,2.0*pi+dphi,dphi)

# Radial critical curve in lens plane
xr=0.0
yr=0.0
```

```python
#calculate critical curves + caustics

# Tangential critical curve in lens plane
xt=[(b*(1.0-gamma*cos(2.0*t))/(1.0-gamma*gamma))*cos(t) for t in phi]
yt=[(b*(1.0-gamma*cos(2.0*t))/(1.0-gamma*gamma))*sin(t) for t in phi]

# Radial caustic in source plane
ur=[-b*cos(t) for t in phi]
vr=[-b*sin(t) for t in phi]

# Tangential caustic in source plane
ut=[-2.0*b*gamma*pow(cos(t),3.0)/(1.0+gamma) for t in phi]
vt=[2.0*b*gamma*pow(sin(t),3.0)/(1.0-gamma) for t in phi]
```

```python
# Plot the critical curves and image positions
subplot(121,aspect='equal')
plot(xr,yr,'ro')
plot(xt,yt,'r-',label='Critical curves')
plot(xi,yi,'bo',label='Images')
grid(True)
xlim(-1.0,1.0)
ylim(-1.0,1.0)
xticks([-1.0,-0.5,0.0,0.5,1.0])
yticks([-1.0,-0.5,0.0,0.5,1.0])
xlabel(r'$x(^{\prime\prime})$')
ylabel(r'$y(^{\prime\prime})$')
legend(title='Lens plane',loc='upper center',bbox_to_anchor=(-0.05,1.05,1.1,0.5),mode='expand',framealpha=1.0,edgecolor='black')

# Plot the caustics and source position
subplot(122,aspect='equal')
plot(ur,vr,'g-')
plot(ut,vt,'g-',label='Caustic')
plot(us,vs,'ko',label='Source')
grid(True)
xlim(-1.0,1.0)
ylim(-1.0,1.0)
xticks([-1.0,-0.5,0.0,0.5,1.0])
yticks([-1.0,-0.5,0.0,0.5,1.0])
xlabel(r'$u(^{\prime\prime})$')
ylabel(r'$v(^{\prime\prime})$')
legend(title='Source plane',loc='upper center',bbox_to_anchor=(-0.05,1.05,1.1,0.5),mode='expand',framealpha=1.0,edgecolor='black')

# Adjust the horizontal spacing and save the figure
subplots_adjust(wspace=0.5)
savefig('SIS_XS.jpg',dpi=300,bbox_inches='tight')
show()
```
