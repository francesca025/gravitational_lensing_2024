# Einstein Ring Patterns from a Point-Like Source behind a Point-Like Lens

Demonstrating the shape of the Einstein Ring arising from viewing a point-like source from behind a point-like gravitational lens.

```python
# import packages
from numpy import *
from pylab import *
rcParams.update({'font.family':'serif','font.size':12})
```

```python
#define data

# Einstein radius
re=1.0

# Source position and distance from coordinate origin
betax=0.3
betay=0.3
beta=sqrt(betax*betax+betay*betay)
```

```python
#compute images

# Position of positive parity image
thetap=(beta+sqrt(beta*beta+4.0*re*re))/2.0
theta1x=thetap*betax/beta
theta1y=thetap*betay/beta

# Position of negative parity image
thetan=(beta-sqrt(beta*beta+4.0*re*re))/2.0
theta2x=thetan*betax/beta
theta2y=thetan*betay/beta

# Einstein ring (needed just for plot)
thex=[re*cos(t) for t in arange(0.0,2.0*pi,pi/180.0)]
they=[re*sin(t) for t in arange(0.0,2.0*pi,pi/180.0)]
```

```python
# Plot the results
ax=subplot(111,aspect='equal')
plot(0,0,'ko',thex,they,'m-',betax,betay,'bo',theta1x,theta1y,'ro',theta2x,theta2y,'go')
grid(True)
xlim(-1.2,1.2)
ylim(-1.2,1.2)
xlabel(r'$x(^{\prime\prime})$')
ylabel(r'$y(^{\prime\prime})$')
text(-0.25,-0.15,'Point-like lens')
text(0.0,0.38,'Point-like source',color='b')
text(0.65,0.97,r'Image$_{1}\ (+)$',color='r')
text(-0.75,-0.47,r'Image$_{2}\ (-)$',color='g')
text(0.55,-0.95,'Einstein ring',color='m')

# Save the figure
savefig('pointl_points.jpg',dpi=300,bbox_inches='tight')
show()
```
