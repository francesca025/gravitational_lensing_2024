# Weak lensing

Demonstration of images from weak lensing. Weak lensing distorts source images and is caused by an extended mass distribution acting as a gravitational lens.

```python
#package imports 

from numpy import *
from matplotlib.pyplot import *
```

```python
#formulas 

# Surface mass density (convergence)
def k(t,re):
    return 0.5*re/t

# Reduced shear
def g(t,re):
    return k(t,re)/(1.0-k(t,re))

# Semi-major and semi-minor axes of an elliptic image
def a(t,re,rs):
    return rs/((1.0-k(t,re))*(1.0-abs(g(t,re))))
def b(t,re,rs):
    return rs/((1.0-k(t,re))*(1.0+abs(g(t,re))))
```

```python
#data

# Einstein radius (in arcmin)
re=0.5

# Source radius (in arcmin)
rs=0.1

# x-coordinates of source center
xs=[0.8,1.0,1.4,1.8]

# Parametrizing variable
dphi=pi/1000.0
phi=arange(dphi,2.0*pi+dphi,dphi)

# Coordinates of elliptic images
x1=[xs[0]+a(xs[0],re,rs)*cos(t) for t in phi]
y1=[b(xs[0],re,rs)*sin(t) for t in phi]
x2=[xs[1]+a(xs[1],re,rs)*cos(t) for t in phi]
y2=[b(xs[1],re,rs)*sin(t) for t in phi]
x3=[xs[2]+a(xs[2],re,rs)*cos(t) for t in phi]
y3=[b(xs[2],re,rs)*sin(t) for t in phi]
x4=[xs[3]+a(xs[3],re,rs)*cos(t) for t in phi]
y4=[b(xs[3],re,rs)*sin(t) for t in phi]
```

```python
# Plot the results and save the figure
subplot(111,aspect='equal')
plot(x1,y1,'r-',x2,y2,'g-',x3,y3,'b-',x4,y4,'m-')
grid(True)
xlim(0.0,2.0)
ylim(-0.5,0.5)
xlabel(r'$x\ (^\prime )$')
ylabel(r'$y\ (^\prime )$')
legend([r'$x_s=0{^\prime}.8$',r'$x_s=1{^\prime}.0$',r'$x_s=1{^\prime}.4$',r'$x_s=1{^\prime}.8$'],loc='upper center',bbox_to_anchor=(0.5,1.2),ncol=2,framealpha=1.0,edgecolor='black')
tight_layout()
savefig('weakl_circs.jpg',dpi=300,bbox_inches='tight')
show()
```
