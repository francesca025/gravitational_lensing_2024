# Convergence and Shear

This notebook computes the convergence $\kappa(\theta)$ and shear $\gamma(\theta)$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def kappa(theta, theta_E):
    return 0.5 * theta_E/theta

def g(theta, theta_E):
    #gamma = kappa
    return kappa(theta, theta_E) / (1 - kappa(theta, theta_E))
```

```python
theta_E = 0.5
```

```python
theta = np.arange(0.5, 3.0, 0.01)
```

```python
kappa_vals = kappa(theta, theta_E)
g_vals = g(theta, theta_E)


plt.plot(theta, kappa_vals, label = r'$\kappa(\theta)$')
plt.plot(theta, g_vals, label = r'$g(\theta)$')
plt.legend()
plt.xlim(0.5, 3.0)
plt.xlabel('$\\theta\\ (^\\prime )$')
plt.legend(['$\\kappa(\\theta)$','$g(\\theta)$'],framealpha=1.0,edgecolor='black')
```
